```python
cov_df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"],inplace= True,axis=1)
```

```python
cov_df.head(5)
```

```python
cov_df['Date']=pd.to_datetime(cov_df['Date'],format = '%Y-%m-%d')
```

```python
cov_df['Date'].head(5)
```

```python
cov_df['Active_case'] = cov_df['Confirmed'] - cov_df['Cured']+cov_df['Deaths']
```

```python
statewise = pd.pivot_table(cov_df,values=['Confirmed','Deaths','Cured'],index = "State/UnionTerritory",aggfunc = max)
```

```python
statewise["recoveryrate"] = statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise["mortalityrate"] = statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise = statewise.sort_values(by = "Confirmed", ascending = False)
```

```python
statewise.style.background_gradient(cmap= 'cubehelix')
```

```python
t10_Act_case = cov_df.groupby(by = 'State/UnionTerritory').max()[['Active_case','Date']].sort_values(by = 'Active_case',ascending = False).reset_index()
```

```python
ax = sns.barplot(data = t10_Act_case.iloc[:10],y = "Active_case",x = "State/UnionTerritory",linewidth = 2,edgecolor = 'red')
```

```python
t10_Act_case = cov_df.groupby(by = 'State/UnionTerritory').max()[['Active_case','Date']].sort_values(by = 'Active_case',ascending = False).reset_index()
fig=plt.figure(figsize=(16,9))
plt.title('Top 10 states with most active cases in India')
ax = sns.barplot(data = t10_Act_case.iloc[:10],y = "Active_case",x = "State/UnionTerritory",linewidth = 2,edgecolor = 'red')
plt.xlabel("States")
plt.ylabel("total active cases")
plt.show()
```

```python
death_case = cov_df.groupby(by = 'State/UnionTerritory').max()[['Deaths','Date']].sort_values(by = 'Deaths',ascending = False).reset_index()
fig=plt.figure(figsize=(16,7))
plt.title('Top 10 states with most death cases in India')
ax = sns.barplot(data = death_case.iloc[:10],y = "Deaths",x = "State/UnionTerritory",linewidth = 2,edgecolor = 'red')
plt.xlabel("States")
plt.ylabel("total Death cases")
plt.show()
```

```python
fig = plt.figure(figsize= (13,7))
ax= sns.lineplot(data = cov_df[cov_df['State/UnionTerritory'].isin(['Maharastra','Karnataka','Kerala','UttarPradesh','TamilNadu'])], x = 'Date', y = 'Active_case', hue = 'State/UnionTerritory')
ax.set_title("top 5 affected states in India")
```
